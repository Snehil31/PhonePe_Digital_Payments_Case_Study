## Task 4: Data Merging and Advanced Analysis
### 4.1: Ratio of users to population by state
 1. Merge the State_Txn and Users dataset with the District Demographics dataset to calculate the ratio of registered users to the population for each state. Display the results in a tabular format.
 2. Create a column chart depicting the ratio of users to population by state.

```python
# Q-4.1: Ratio of users to population by state
# 1 - Merge the State_Txn and Users dataset with the District Demographics dataset to calculate the ratio of registered users to the population for each state. Display the results in a tabular format.

state_group = State_Txn_Users_df.groupby(['State'])['Registered Users'].sum().reset_index()

demo_group = District_Demographics_df.groupby(['State'])['Population'].sum().reset_index()

user_to_pop = state_group.merge(demo_group, on=['State'])

user_to_pop['user_to_pop_ratio'] = user_to_pop['Registered Users'] / user_to_pop['Population']
print(user_to_pop)


# Q-4.1: Ratio of users to population by state
# 2 - Create a column chart depicting the ratio of users to population by state.

plt.figure(figsize=(15, 4))
sns.barplot(data=user_to_pop.sort_values('user_to_pop_ratio', ascending=False), x="State", y="user_to_pop_ratio")
plt.xticks(rotation=90)
plt.title(f"Ratio of users to population by state")
plt.show()
```

### 4.2: Correlate population density with transaction volume
 1. Merge the District_Txn and Users dataset with the District Demographics dataset.
 2. Calculate the correlation between population density and transaction volume.
 3. Create a scatter plot to visualize the correlation between population density and transaction volume.

```python
# Q-4.2: Correlate population density with transaction volume
# 1 - Merge the District_Txn and Users dataset with the District Demographics dataset.

pop_to_trans = District_Txn_Users_df.merge(District_Demographics_df, on=['State', 'District', 'Code'])

district_transactions = pop_to_trans.groupby(["State","District"]).agg({
    "Transactions":"sum",
    "Density":"max"
}).reset_index()


# Q-4.2: Correlate population density with transaction volume
# 2 - Calculate the correlation between population density and transaction volume.

pop_to_trans_correl = district_transactions['Density'].corr(district_transactions['Transactions'])
print(f'Correlation between population density and transaction volume: {pop_to_trans_correl}')


# Q-4.2: Correlate population density with transaction volume
# 3 - Create a scatter plot to visualize the correlation between population density and transaction volume.

plt.figure(figsize=(15, 4))
sns.scatterplot(data=pop_to_trans, x='Density', y='Transactions', hue='State')
plt.title('Population Density vs Transaction Volume')
plt.legend(loc='upper right', ncol=5)
plt.show()
```

### 4.3: Average transaction amount per user
 1. Merge relevant datasets to calculate the average transaction amount per user for each state. Display the results in a tabular format.
 2. Identify the top 5 states with the highest average transaction amount per user and the top 5 states with the lowest average transaction amount per user. Display the results.

```python
# Q-4.3: Average transaction amount per user
# 1 - Merge relevant datasets to calculate the average transaction amount per user for each state. Display the results in a tabular format.

trans_users_data = State_Txn_Users_df.groupby(['State']).agg({
    'Amount (INR)':'sum',
    'Registered Users':'sum'
}).reset_index()

trans_users_data['avg_trans_per_user'] = trans_users_data['Amount (INR)'] / trans_users_data['Registered Users']
print(f'Average transaction amount per user for each state: \n {trans_users_data[['State', 'avg_trans_per_user']]}')
```

```python
# Q-4.3: Average transaction amount per user
# 2 - Identify the top 5 states with the highest average transaction amount per user and the top 5 states with the lowest average transaction amount per user. Display the results.

largest_5_avg_trans = trans_users_data.nlargest(5, 'avg_trans_per_user')[['State', 'avg_trans_per_user']]
print(f'Top 5 states with the highest average transaction amount per user: \n {largest_5_avg_trans}')
print()

lowest_5_avg_trans = trans_users_data.nsmallest(5, 'avg_trans_per_user')[['State', 'avg_trans_per_user']]
print(f'Top 5 states with the lowest average transaction amount per user: \n {lowest_5_avg_trans}')
```

### 4.4: Device brand usage ratio
 1. Merge the State_DeviceData dataset with the State_Txn and Users dataset.
 2. Calculate the ratio of users using each device brand to the total number of registered users in each state. Display the results in a tabular format.
 3. Create a bar chart depicting the device brand usage ratio for each state.

```python
# Q-4.4: Device brand usage ratio
# 1 - Merge the State_DeviceData dataset with the State_Txn and Users dataset.
# 2 - Calculate the ratio of users using each device brand to the total number of registered users in each state. Display the results in a tabular format.

merg_data = State_Txn_Users_df[['State', 'Registered Users']].merge(State_Device_Data_df[['State', 'Registered Users', 'Brand']], on=['State'], suffixes=['_total', '_brand'])
merged_device = merg_data.groupby(['State', 'Brand'])[['Registered Users_brand', 'Registered Users_total']].sum().reset_index()
merged_device['user_ratio'] = merged_device['Registered Users_brand'] / merged_device['Registered Users_total']
print(f'Ratio of users using each device brand to the total number of registered users in each state: \n {merged_device[['State', 'Brand', 'user_ratio']]}')


# Q-4.4: Device brand usage ratio
# 3 - Create a bar chart depicting the device brand usage ratio for each state.

plt.figure(figsize=(15, 4))
sns.barplot(data=merged_device, x='State', y='user_ratio', hue='Brand')
plt.xticks(rotation=90)
plt.title("Device brand usage ratio per state")
plt.legend(loc='upper right', ncol=9)
plt.show()
```
